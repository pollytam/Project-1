```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
crime_data = pd.read_csv("./Police_Department_Incident_Reports__2018_to_Present.csv")
crime_df=pd.DataFrame(crime_data, columns=["Incident ID","Incident Category", "Incident Time", "Incident Day of Week"])
crime_df.head()
```

```python
crime_df['Incident Category'] = crime_df['Incident Category'].replace(
    {'Embezzlement': 'Burglary', 'Larceny Theft': 'Burglary','Lost Property': 'Burglary',
     'Vehicle Misplaced': 'Burglary', 'Motor Vehicle Theft': 'Burglary','Motor Vehicle Theft?': 'Burglary',
     'Robbery': 'Burglary', 'Stolen Property': 'Burglary',
     
     'Recovered Vehicle': 'Case Closure',
     'Malicious Mischief': 'Courtesy Report','Disorderly Conduct': 'Courtesy Report', 'Assault': 'Courtesy Report',
     'Civil Sidewalks': 'Courtesy Report', 'Vandalism': 'Courtesy Report',
     
     'Drug Violation': 'Drug Offense','Liquor Laws': 'Drug Offense',
     
     'Offences Against The Family And Children': 'Family Offense','Arson': 'Fire Report', 
     
     'Human Trafficking (A), Commercial Sex Acts': 'Prostitution','Human Trafficking, Commercial Sex Acts': 'Prostitution',
     
     'Forgery And Counterfeiting': 'Fraud', 
     
     'Other Miscellaneous': 'Miscellaneous','Other Offenses': 'Miscellaneous',
     'Miscellaneous Investigation': 'Miscellaneous','Other': 'Miscellaneous', 'Non-Criminal': 'Miscellaneous',
     
     'Rape': 'Sex Offense',
     
     'Suspicious Occ': 'Suspicious', 
     
     'Traffic Collision': 'Traffic Collision/Violation','Traffic Violation Arrest': 'Traffic Collision/Violation',
     'Vehicle Impounded': 'Traffic Collision/Violation', 
     
     'Weapons Offence': 'Weapons Offense','Weapons Carrying Etc': 'Weapons Offense'
    })


crime_df.head()
```

```python
#CATEGORY VS COUNT
incidentCategory = "Incident Category"
incidentID = "Incident ID"
#Set Up DataFrame
incidentCountDF = crime_df.loc[:, [incidentCategory,incidentID ]].groupby(incidentCategory, as_index=False).count()
incidentCountDF = incidentCountDF.rename(columns={incidentID: "Incident Count"})
incidentCountDF = incidentCountDF.set_index('Incident Category')

#EXAMPLE
#EXAMPLE
#EXAMPLE
#EXAMPLE

# https://stackoverflow.com/questions/47805767/why-does-pyplot-plot-create-an-additional-rectangle-with-width-1-height-1/47812553#47812553

#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.bar.html

#https://matplotlib.org/examples/api/barchart_demo.html

#DISPLAY VERTICALLY
#DISPLAY VERTICALLY
#DISPLAY VERTICALLY
#DISPLAY VERTICALLY

incidentCountDF = incidentCountDF.sort_values(by="Incident Count")
ax = incidentCountDF.plot(kind='bar',figsize=(10,10))
def autolabel(rects, ax):
    for rect in rects:
        x = rect.get_x() + rect.get_width()/2.
        y = rect.get_height()
        ax.annotate("{}".format(y), (x,y), xytext=(0,5), textcoords="offset points",
                    ha='center', va='bottom', rotation=90)
autolabel(ax.patches,ax)
#adjust for long bars
ax.margins(y=0.5)
#adjust for long text
plt.subplots_adjust(bottom=0.50)
plt.title('SF Incidents in 2018')
plt.ylabel('Incident Count')
plt.xlabel("Incident Type")
plt.show()



#DISPLAY HORIZONTALLY
#DISPLAY HORIZONTALLY
#DISPLAY HORIZONTALLY
#DISPLAY HORIZONTALLY
#DISPLAY HORIZONTALLY
# incidentCountDF = incidentCountDF.sort_values(by="Incident Count")
# ax = incidentCountDF.plot(kind='barh')
# def autolabel(rects):
#     for rect in rects:
#         width = rect.get_width()
#         plt.text(1.15*rect.get_width(), rect.get_y()+0.5*rect.get_height(),
#                     '%d' % int(width),
#                     ha='center', va='center')
# autolabel(ax.patches)
# ax.margins(y=0.1)
# plt.ylabel('Incident Count')
# plt.xlabel("Incident Type")
# plt.show()
plt.savefig("total_crime_bar.png")
```

```python
incidentDayOfWeek  = "Incident Day of Week"
incidentCategory = "Incident Category"
incidentTime = "Incident Time"
#GET TOP 10 CRIMES
incidentDF = crime_df.loc[:, [incidentDayOfWeek,incidentCategory ]].groupby(incidentCategory, as_index=False).count().sort_values(by=incidentDayOfWeek, ascending=False).head(10)
incidentDF = incidentDF.rename(columns={incidentDayOfWeek: "Incident Count"})
# print(incidentDF)
bins = [0,100,200,300,400,500,600,700,800,900,1000,1100,1200,1300,1400,1500,1600,1700,1800,1900,2000,2100,2200,2300,2400]
##bins = [0,400,800,1200,1600,2000,2400]
timeOfDayDF = crime_df.loc[:, [incidentTime,incidentCategory ]]
timeOfDayDF["Count Per Day"] = timeOfDayDF.groupby([incidentTime,incidentCategory])[incidentCategory].transform("count")
finalDF = pd.merge(timeOfDayDF,incidentDF, on=incidentCategory).drop_duplicates()
# CONVERT COLUMN TO INT SO I CAN BIN
def convertToInt(x):
    temp = x.split(":")
    tempString = "".join(temp)
    return int(tempString)
finalDF[incidentTime] = finalDF[incidentTime].apply(convertToInt)
#binning 
finalDF['binned'] = pd.cut(finalDF[incidentTime], bins)
# print(finalDF)
#listOfTopCrimes
#listOfTopCrimes
#listOfTopCrimes
top10IncidentsList = finalDF[incidentCategory].unique()
#SORT AND GROUP BY CATEGORY
for incident in top10IncidentsList:
    print(incident)
    binnedDF = finalDF.loc[:, ["binned", incidentCategory]]
    binnedDFSorted = binnedDF.loc[binnedDF[incidentCategory] == incident].groupby("binned", as_index=False).count()
    binnedDFSorted = binnedDFSorted.rename(columns={incidentCategory:"Incident count during time"})
    listToPlot = binnedDFSorted["Incident count during time"].tolist()
    plt.plot(listToPlot, label = incident)
    # print("INCIDENT: " + incident)
    # print(binnedDFSorted)
plt.legend(bbox_to_anchor=(1.001, 1), loc=2, borderaxespad=0.)
plt.xticks(np.arange(24), ['0','100','200','300','400','500','600','700','800','900','1000','1100','1200','1300','1400','1500','1600','1700','1800','1900','2000','2100','2200','2300','2400'], rotation=90)
#plot.set_xticklabels(binnedDF["binned"], rotation=0)
#plt.xticks([0-400],'400-800','800-1200','1200-1600','1600-2000','2000-2400'])
##plt.xticks(np.arange(4), ['0-600','600-1200','1200-1800','1800-2400'])
plt.ylabel('Incident Count')
plt.xlabel("Time")
plt.title("Incident - Time of a Day")
plt.show()
```
